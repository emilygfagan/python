```python
#identifying the columns that have null values
df.isnull().sum()
```

### The column of interest that has null values is Credit_Limit. I am going to take the median credit limit and impute it into the missing values to fix the problem. Since there are only 60 null values out of the whole dataset, there are many outliers, and the median is less than the mean, this is the best option.

```python
#filling null values with the median Credit_limit
df['Credit_Limit'] = df['Credit_Limit'].fillna(df['Credit_Limit'].median())
```

```python
#checking to see if there are any null values left
df.isnull().sum()
```
